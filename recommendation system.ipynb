```python
def convert(obj):
    li=[]
    for i in ast.literal_eval(obj):
        li.append(i["name"])
    return li
# print(li)
```

```python
movies["genres"]=movies["genres"].apply(convert)
```

```python
def convertx(obj):
    li=[]
    cou=0
    for i in ast.literal_eval(obj):
        if cou!=3:
            li.append(i["name"])
            cou+=1
            
        else:
            break
    return li
```

```python
movies["cast"]=movies["cast"].apply(convertx)
```

```python
movies["crew"].values
```

```python
def fetch_dectector(ojs):
    le=[]
    for i in ast.literal_eval(ojs):
        if i["job"]=="Director":
            le.append(i["name"])
            break
    return le
```

```python
movies["crew"]=movies["crew"].apply(fetch_dectector)
```

```python
movies["overview"]=movies["overview"].apply(lambda x: x.split())

# def rep(ojb2):
#     for i in ojb2:
#         i.replace(" ","")
#     return i
```

```python
movies["cast"]=movies["cast"].apply(lambda x: [i.replace(" ","") for i in x])
movies["crew"]=movies["crew"].apply(lambda x: [i.replace(" ","") for i in x])
movies["genres"]=movies["genres"].apply(lambda x: [i.replace(" ","") for i in x])
movies["keywords"]=movies["keywords"].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
movies.head()
```

```python
new_movies = movies[["movie_id","title","tags"]]
new_movies
```

```python
new_movies["tags"]=new_movies["tags"].apply(lambda x:" ".join(x))
new_movies["tags"]
```

```python
new_movies.head()
```

```python
new_movies["tags"]=new_movies["tags"].apply(lambda x: x.lower())
```

```python
new_movies.head()
```

```python

import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# ps.stem("dancing")
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_movies["tags"]=new_movies["tags"].apply(stem)
```

```python
new_movies["tags"][0]
```

```python
new_movies["tags"][1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(new_movies["tags"]).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity
```

```python
sorted(similarity[0])[-10:-1]
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x: x[1]) [1:6]
```

```python
enumerate(similarity[0])
```

```python
def recommend(movie):
    movie_index=new_movies[new_movies["title"]==movie].index[0]
    dis=similarity[movie_index]
    movies_list=sorted(list(enumerate(dis)), reverse=True, key=lambda x: x[1]) [1:6]    

    for i in movies_list:
        print(new_movies.iloc[i[0]].title)
```

```python
recommend("Avatar")
```
